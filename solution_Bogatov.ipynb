```python
idxmax = (data['Total Interactions'] / data['Lifetime Post Total Reach']).idxmax()
```

```python
data.iloc[idxmax]
```

Самый популярный объект в выборке - объект под номером 87. 
Популярность высчитывал по формуле: max(Total Interactions / Lifetime Post Total Reach).
Выбрал такую формулу, потому что Total Interactions это сумма лайков, комментариев и репостов, а Lifetime Post Total Reach это количество уникальных пользователей, просмотревших пост.
